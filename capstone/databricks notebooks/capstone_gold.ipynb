```python
df = spark.read.table("workspace.default.silver_exports")
```

```python
from pyspark.sql.functions import sum, avg

country_year = (
    df.groupBy("country_code", "country_name", "region", "year")
      .agg(
          sum("export_quantity").alias("total_quantity"),
          sum("export_value_usd").alias("total_value_usd"),
          avg("unit_price").alias("avg_unit_price")
      )
)

country_year.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("workspace.default.gold_country_year")
```

```python
product_year = (
    df.groupBy("product_code", "product_name", "year")
      .agg(
          sum("export_quantity").alias("total_quantity"),
          sum("export_value_usd").alias("total_value_usd")
      )
)

product_year.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("workspace.default.gold_product_year")
```

```python
region_year = (
    df.groupBy("region", "year")
      .agg(
          sum("export_value_usd").alias("total_value_usd")
      )
)

region_year.write \
    .mode("overwrite") \
    .format("delta") \
    .saveAsTable("workspace.default.gold_region_year")
```
